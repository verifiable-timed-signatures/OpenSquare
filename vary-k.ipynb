```python
import matplotlib.pyplot as plt 
```

```python
x = [i for i in range(1,20)]
ask_costs = [99395,  93373, 104782,
  115468, 126154, 136841,
  147527, 158213, 168899,
  179585, 190272, 200958,
  211644, 222330, 233017,
  243703, 254389, 265076,
  275762]
```

```python
plt.rc('text',usetex=True)

plt.rcParams["figure.figsize"] =(5, 3)
plt.rcParams.update({'font.size': 13})

ask_costs = [i/1000 for i in ask_costs]

plt.plot(x,ask_costs, linestyle='--', label="Ask Function")

plt.legend()
plt.xlabel("Number of solvers ($k$)")
plt.ylabel("Worst case Gas costs (in kgas)")
plt.savefig("gas-cost-variation.pdf", bbox_inches="tight")
```
